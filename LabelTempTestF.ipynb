```python
# 7 channels

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import csv
import time
from IPython.display import clear_output

img_path = 'Diagram.png'
img = mpimg.imread(img_path)

def lastest_data(csv_file):
    with open(csv_file, 'r') as file:
        data = csv.reader(file)
        latest_entry = None
        for row in data:
            latest_entry = row 
        return latest_entry

def update_image(image_path, CH1, CH2,CH3,CH4,CH5,CH6,CH7, timestamp):
    img = mpimg.imread(image_path)
    plt.figure(dpi=200)
    plt.imshow(img)

    plt.text(430, 420, f"CH1: {CH1} C", fontsize=8, color='black', fontweight='bold',)
    plt.scatter(390,350,color='black',edgecolors='white')
    
    plt.text(450, 780, f"CH2: {CH2} C", fontsize=8, color='white', fontweight='bold')
    plt.scatter(500,820,color='white',edgecolors='black')
    
    plt.text(470, 1170, f"CH3: {CH3} C", fontsize=8, color='red', fontweight='bold')
    plt.scatter(470,1100,color='red',edgecolors='white')
        
    plt.text(300,1470, f"CH4: {CH4} C", fontsize=8, color='lime', fontweight='bold')
    plt.scatter(300,1400,color='lime',edgecolors='white')

    plt.text(1750,730, f"CH5: {CH5} C", fontsize=8, color='magenta', fontweight='bold')
    plt.scatter(1750,660,color='magenta',edgecolors='white')
    
    plt.text(1500,1270, f"CH6: {CH6} C", fontsize=8, color='cyan', fontweight='bold')
    plt.scatter(1500,1200,color='cyan',edgecolors='white')
    
    plt.text(1140,1470, f"CH7: {CH7} C", fontsize=8, color='blue', fontweight='bold')
    plt.scatter(1140,1400,color='blue',edgecolors='white')
    
    
    plt.title(timestamp)  
    plt.show()

csv_file = 'realtimedata_test.csv'

latest_entry = lastest_data(csv_file)
if latest_entry:
    timestamp = str(latest_entry[0]) 
    CH1 = float(latest_entry[1]) 
    CH2 = float(latest_entry[2]) 
    CH3 = float(latest_entry[3])
    CH4 = float(latest_entry[4])
    CH5 = float(latest_entry[5])
    CH6 = float(latest_entry[6])
    CH7 = float(latest_entry[7])
    
    update_image(img_path, CH1, CH2, CH3,CH4,CH5,CH6,CH7, timestamp)
    
    
if __name__ == "__main__":
    try:
        while True:
            time.sleep(1) 
            new_entry = lastest_data(csv_file)
            if new_entry != latest_entry:
                latest_entry = new_entry
                timestamp = str(latest_entry[0]) 
                CH1 = float(latest_entry[1]) 
                CH2 = float(latest_entry[2]) 
                CH3 = float(latest_entry[3])
                CH4 = float(latest_entry[4])
                CH5 = float(latest_entry[5])
                CH6 = float(latest_entry[6])
                CH7 = float(latest_entry[7])
        
                clear_output(wait=True) 
                update_image(img_path, CH1, CH2,CH3,CH4,CH5,CH6,CH7, timestamp)
                
    except KeyboardInterrupt:
            print("stop")
```
